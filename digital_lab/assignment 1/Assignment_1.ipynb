```python
# Path to image in Google Drive
image_path = "/content/drive/MyDrive/UITS/Eight Semester/Digital Image Processing/Assigments/Class 1/pizza.jpg"

# Task 1: Extract a 100Ã—100 Region from the Center of the Image

def extract_center_region(image_path):
    image = np.array(Image.open(image_path))

    h, w, _ = image.shape
    x, y = w // 2, h // 2
    cropped_region = image[y-50:y+50, x-50:x+50]

    plt.imshow(cropped_region)
    plt.axis("off")
    plt.title("100x100 Center Region")
    plt.show()

    return cropped_region
```

```python
# Task 2: Convert the image to HSV and show components

def convert_to_hsv(image_path):
    image = np.array(Image.open(image_path))
    hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

    h, s, v = cv2.split(hsv_image)

    for title, component in zip(["Hue", "Saturation", "Value"], [h, s, v]):
        plt.imshow(component, cmap='gray')
        plt.axis("off")
        plt.title(title)
        plt.show()

    return hsv_image
```

```python
# Run the functions
cropped_image = extract_center_region(image_path)
hsv_image = convert_to_hsv(image_path)
```
